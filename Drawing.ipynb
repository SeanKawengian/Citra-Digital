```python
import cv2
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r"D:\Skyrim.png")
```

```python
plt.imshow (img)
```

```python
cv2.rectangle (img, pt1 = (800, 470), pt2 = (980, 530), color = (255, 0, 0), thickness = 5)
plt.imshow(img)
```

```python
cv2.circle(img, center = (950, 50), radius = 50,color = (0, 0, 255), thickness = 5) 
plt.imshow(img)
```

```python
cv2.putText(img, text = "Skyrim",org = (250, 250),fontFace = cv2.FONT_HERSHEY_DUPLEX, fontScale = 2, color = (0, 255, 0), thickness = 2, lineType = cv2.LINE_AA) 
plt.imshow(img)
```
